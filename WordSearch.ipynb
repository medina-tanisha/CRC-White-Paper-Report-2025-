```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = list(stopwords.words('english'))


custom_stopwords = ['lot', 'really', 'new', 'us', 'could', 'things', 'katy', 'sarah',
                    'anything', 'else', 'anything else', 'like', 'add', 'also', 'would', 'notes', 'take', 'colorado', 'one', 'question', 'yet', '4did', 'need']


all_stopwords = stop_words + custom_stopwords


from google.colab import files
uploaded = files.upload()
file_name = next(iter(uploaded))

df = pd.read_excel(file_name)


columns_to_analyze = [
    'Needs',
    '1. Most appreciate about working with GrantCorps?',
    '2. Did we help you address the grant development needs we identified?.1.2',
    '3. What activities and services provided were most helpful?',
    '4. Any NEW grants?',
    '5. How can we improve program?',
    '6. Anything else you would like us to know?'
]


for col in columns_to_analyze:
    print(f"\nðŸ”¹ Top results for column: {col}")

    text = df[col].dropna().astype(str).str.lower().str.replace('[^\w\s]', '', regex=True)

    vec = CountVectorizer(stop_words=all_stopwords)
    X = vec.fit_transform(text)
    word_freq = pd.DataFrame(X.toarray(), columns=vec.get_feature_names_out())
    word_counts = word_freq.sum().sort_values(ascending=False)
    print("Top words:")
    print(word_counts.head(10))

    vec2 = CountVectorizer(ngram_range=(2, 2), stop_words=all_stopwords)
    X2 = vec2.fit_transform(text)
    bigram_freq = pd.DataFrame(X2.toarray(), columns=vec2.get_feature_names_out())
    bigram_counts = bigram_freq.sum().sort_values(ascending=False)
    print("\nTop 2-word phrases:")
    print(bigram_counts.head(5))

    print("\n" + "-"*50)
```
